Prediciting Stock Prices


##Packages installed

```python
import csv
import numpy as np
from sklearn.svm import SVR
import pandas as pd
import matplotlib.pyplot as plt.
import xgboost as xgb
from sklearn.model_selection import GridSearchCV, TimeSeriesSplit
from sklearn.metrics import mean_squared_error
```

##Data
**helper function**

```python
get_data('appl.csv')
print(dates)
```

```python
# creating dataframes

dates = []
prices = []

#helper function to load data prices as float and dates as int

def get_data(filename):
    with open(filename, 'r') as csvfile:
      csvFileReader = csv.reader(csvfile)
      next(csvFileReader)
      for row in csvFileReader:
          dates.append(int(row[0].split('/')[1]))
          # Remove the '$' sign before converting to float
          prices.append(float(row[3].replace('$', '')))
    return
```

##XGBoost model


```python
params = {
    'min_child_weight': [1, 5, 10],
    'gamma': [0.5, 1, 1.5, 2, 5],
    'subsample': [0.6, 0.8, 1.0],
    'colsample_bytree': [0.6, 0.8, 1.0],
    'max_depth': [3, 4, 5]
      }
```

***grid search***

```python
xgbr = xgb.XGBRegressor(objective='reg:squarederror') # You can specify other objective functions

dates_2d = np.array(dates).reshape(-1, 1)
prices_2d = np.array(prices).reshape(-1, 1)

grid_search = GridSearchCV(xgbr, param_grid=params, cv=5, scoring='neg_mean_squared_error', verbose=True)
grid_search.fit(dates_2d, prices_2d)

print(grid_search.best_params_)
```

```python
best_params = grid_search.best_params_
best_score = grid_search.best_score_
```

```python
best_model = grid_search.best_estimator_
```

##SVR Model

```python
def pred_prices(dates , prices, x):
  dates = np.reshape(dates,(len(dates),1))
  svr_lin = SVR(kernel='linear', C=1e3)
  svr_poly = SVR(kernel= 'poly', C=1e3, degree= 2)
  svr_rbf = SVR(kernel='rbf', C=1e3, gamma=0.1)

  svr_lin.fit(dates, prices)
  svr_poly.fit(dates, prices)
  svr_rbf.fit(dates, prices)

  plt.scatter(dates, prices, color = 'black', label= "Data")
  plt.plot(dates, svr_rbf.predict(dates), color= 'red', label= 'RBF model')
  plt.plot(dates, svr_poly.predict(dates), color= 'green', label= 'Polynomial model')
  plt.plot(dates, svr_lin.predict(dates), color= 'blue', label= 'Linear model')
  plt.plot(dates, best_model.predict(dates), color= 'pink', label= 'XGBoost')
  plt.xlabel('Date')
  plt.ylabel('Price')
  plt.title('Support Vector Regression')
  plt.legend()
  plt.show()

  # Reshape x to be a 2D array
  x = np.array(x).reshape(-1, 1)

  return svr_rbf.predict(x)[0], svr_poly.predict(x)[0], svr_lin.predict(x)[0], best_model.predict(x)[0]
```

```python
predictedprice = pred_prices(dates, prices, 29)
print(predictedprice)
```
